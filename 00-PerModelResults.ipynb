```python
%matplotlib inline
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import os

RESULTS_PATH = 'results/weather/'
NOTEBOOKS = ['01-Basic',
             '02-Chunk1', '02-Chunk2', '02-Chunk3', '02-Chunk4', '02-Chunk5',
             '03-FeaturesDate', '03-FeaturesDate_n', '03-FeaturesDate_p', '03-FeaturesDate_pn',
             '03-FeaturesOther', '03-FeaturesOther_n', '03-FeaturesOther_p', '03-FeaturesOther_pn',
             '04-Remove20_n', '04-Remove30_n', '04-Remove40_n', '04-Remove50_n', '04-Remove60_n', '04-Remove70_n', '04-Remove80_n']
```

```python
# Bar chart for a single model
def plot_bar(model):
    df_all = pd.DataFrame()
    
    for notebook in NOTEBOOKS:
        df = pd.read_csv(RESULTS_PATH + notebook + '.csv')
        append_object = {
            'notebook': notebook,
            'train': df[df['model'] == model]['train'].values[0],
            'test': df[df['model'] == model]['test'].values[0],
        }
        df_all = df_all.append(append_object, ignore_index=True)
        

    df_all.plot(x='notebook', y=['train', 'test'], kind='bar', title=model, figsize=(15, 10), legend=True, fontsize=12)
    plt.ylabel('RMSE Loss (Root Mean Squared Error)')
    plt.show()
    
    df_all.sort_values(['test'], inplace=True)
    
    print(df_all)
```
