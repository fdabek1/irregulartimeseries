Looking at the severity of each condition over all windows.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('fivethirtyeight')
```

```python
import json

with open('data/targets_severity/stats.json') as f:
    stats = json.load(f)
    
codes = stats['codes']
```

```python
def draw_chart(index):
    y = np.load('data/targets_severity/y.npy')
    y = y[:, :, index]

    target_windows = [[0, 30], [30, 60], [60, 90], [90, 120], [120, 150], [150, 180], [180, 210], [210, 240],
                      [240, 270],
                      [270, 300], [300, 330], [330, 360]]
    window_strings = [str(window[0]) + '-' + str(window[1]) for window in target_windows]
    window_totals = y.sum(axis=0)
    
    n_windows = np.arange(len(target_windows))
    for w in n_windows:
        print(window_strings[w] + ':		' + str(window_totals[w]))

    plt.bar(n_windows, window_totals, align='center',
            alpha=1.0, color='b')

    plt.xlabel('Windows')
    plt.ylabel('Counts')
    plt.title('Severity per Window')
    plt.xticks(n_windows, window_strings, rotation=-90)

    plt.show()
```

```python
from IPython.display import display, HTML

for c, code in enumerate(codes):
    display(HTML('<h3>' + code + '</h3>'))
    draw_chart(c)
```
